```python
import matplotlib.pyplot as plt

# Define the years and market capitalization values for each company
years = ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']

tesla_market_cap = [0, 0, 4.33, 18.15, 26.59, 30.45, 34.66, 52.97, 52.06, 76.03, 669.93, 752.54]
apple_market_cap = [283.42, 372.64, 481.89, 454.12, 615.51, 651.63, 616.37, 814.33, 745.94, 1200, 2030, 2430]
microsoft_market_cap = [201.61, 205.37, 231.29, 257.95, 380.95, 416.06, 442.3, 652.18, 781.88, 1050, 1390, 2140]
nvidia_market_cap = [8.9, 13.57, 7.73, 8.98, 11.37, 13.92, 32.35, 107.77, 133.04, 121.46, 325.18, 505.69]

# Set the dark background style
plt.style.use('dark_background')

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting the market capitalization comparison
ax.plot(years, tesla_market_cap, label='Tesla', marker='o')
ax.plot(years, apple_market_cap, label='Apple', marker='o')
ax.plot(years, microsoft_market_cap, label='Microsoft', marker='o')
ax.plot(years, nvidia_market_cap, label='NVIDIA', marker='o')

# Set the plot title and labels
ax.set_title('Market Capitalization Comparison')
ax.set_xlabel('Year')
ax.set_ylabel('Market Cap (Billions USD)')

# Set the legend
ax.legend()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# Define the stock symbols for Tesla, Microsoft, Apple, and Nvidia
symbols = ['TSLA', 'MSFT', 'AAPL', 'NVDA']

# Download historical stock data from Yahoo Finance
data = yf.download(symbols, start='2012-01-01', end='2021-12-31', group_by='ticker', auto_adjust=True)

# Plotting the stock prices
plt.figure(figsize=(12, 6))

for symbol in symbols:
    column_name = data[symbol].columns[0]  # Get the first column name
    plt.plot(data.index, data[symbol][column_name], label=symbol)

# Set the plot title and labels
plt.title('Stock Prices Comparison')
plt.xlabel('Date')
plt.ylabel('Price (USD)')

# Set the legend
plt.legend()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# Define the stock symbols for Tesla, Microsoft, Apple, and Nvidia
symbols = ['TSLA', 'MSFT', 'AAPL', 'NVDA']

# Download historical stock data from Yahoo Finance
data = yf.download(symbols, start='2012-01-01', end='2021-12-31', group_by='ticker', auto_adjust=True)

# Resample the data to monthly frequency and calculate the monthly percentage change of stock prices
monthly_returns = data.resample('M').ffill().pct_change()

# Plotting the monthly percentage change of stock prices
plt.figure(figsize=(12, 6))

for symbol in symbols:
    plt.plot(monthly_returns.index, monthly_returns[symbol]['Close'], label=symbol)

# Set the plot title and labels
plt.title('Monthly Stock Price Percentage Change Comparison')
plt.xlabel('Date')
plt.ylabel('Monthly Percentage Change')
plt.legend()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# Define the stock symbols for Tesla, Microsoft, Apple, and Nvidia
symbols = ['TSLA', 'MSFT', 'AAPL', 'NVDA', 'AMD']

# Download historical stock data from Yahoo Finance
data = yf.download(symbols, start='2012-01-01', end='2021-12-31', group_by='ticker', auto_adjust=True)

# Calculate the cumulative percentage change of stock prices
cumulative_returns = (data / data.iloc[0] - 1) * 100

# Plotting the cumulative percentage change of stock prices
plt.figure(figsize=(12, 6))

for symbol in symbols:
    plt.plot(cumulative_returns.index, cumulative_returns[symbol]['Close'], label=symbol)

# Set the plot title and labels
plt.title('Cumulative Stock Price Growth Comparison')
plt.xlabel('Date')
plt.ylabel('Cumulative Percentage Change')
plt.legend()

# Display the plot
plt.show()
```
